### Key Columns Identified

- booking_status: Indicates whether the ride was completed, cancelled, or incomplete
- vehicle_type: Type of vehicle used for the ride
- payment_method: Mode of payment used by customer
- booking_value: Total fare amount for the ride
- ride_distance: Distance covered during the ride
- customer_rating: Rating given by customer
- driver_rating: Rating given to driver
- cancellation_reason: Reason for cancellation (customer or driver)
