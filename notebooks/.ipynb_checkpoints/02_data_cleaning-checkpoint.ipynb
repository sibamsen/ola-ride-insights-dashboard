```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../data/raw/OLA_DataSet.xlsx")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Canceled_Rides_by_Customer'] = df['Canceled_Rides_by_Customer'].fillna('No')
df['Canceled_Rides_by_Driver'] = df['Canceled_Rides_by_Driver'].fillna('No')
```

```python
df['Incomplete_Rides'] = df['Incomplete_Rides'].fillna('No')
df['Incomplete_Rides_Reason'] = df['Incomplete_Rides_Reason'].fillna('Not Applicable')
```

```python
# Ratings remain NaN where rides were not completed
df['Driver_Ratings'] = df['Driver_Ratings']
df['Customer_Rating'] = df['Customer_Rating']
```

```python
df['Payment_Method'] = df['Payment_Method'].fillna('Unknown')
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
```

```python
df['Ride_Day'] = df['Date'].dt.day_name()
df['Ride_Month'] = df['Date'].dt.month
df['Ride_Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.hour
```

```python
df.isnull().sum()
```

```python
df.to_csv("../data/processed/ola_cleaned.csv", index=False)
```
